# To add a column for which team bat first

```python
def bat_first_team(t_d,t_w,t1,t2):
    if t_d=="bat":
        return t_w
    else:
        k = set()
        k.add(t1)
        k.add(t2)
        k.remove(t_w)
        return list(k)[0]

iplhistory["bat_first"]=iplhistory.apply(lambda x: bat_first_team(x["toss_decision"],x["toss_winner"],x["team1"],x["team2"]),axis=1)
#iplhistory["bat_first"] = np.where(iplhistory["toss_decision"]=="bat",iplhistory["toss_winner"],)

#if result is by runs then winner is the second batting
#if result is by wickets then the first batting is the winner of the match
#here use a lambda and apply function by using a function to extract the row required by writing the condition in the function
iplhistory
```
